```python
fp = open('1661-0.txt', 'r')
```

```python
file_buffer = fp.read()
```

```python
type(file_buffer)
```

```python
help(str)
```

```python
len(file_buffer)
```

```python
file_buffer
```

```python
file_list = file_buffer.split()
```

```python
type(file_list)
```

```python
file_list
```

```python
import collections as cc
```

```python
file_collection = cc.Counter(file_list)
```

```python
type(file_collection)
```

```python
file_collection
```

```python
file_dict = dict(file_collection)
```

```python
file_dict
```

```python
file_word_features = [(i, j) for i,j in file_dict.items() if (len(i) > 5) and j > 10 and j < 20]
```

```python
file_word_features
```

```python
file_feature_list = str(file_word_features)
```

```python
fp2 = open('feature_list.txt', 'w')
```

```python
fp2.write(file_feature_list)
```

```python
fp.close()
fp2.close()
```
