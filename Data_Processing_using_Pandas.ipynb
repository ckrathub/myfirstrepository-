```python
import numpy as np
import pandas as pd
```

```python
stocks = pd.Series(np.random.randn(1000))
```

```python
stocks.index=range(1000, 2000, 1)
```

```python
stocks.cumsum().plot()
```

```python
stocks = pd.DataFrame(np.random.randn(1000, 4))
```

```python
stocks.index = pd.date_range('1/1/2020', periods=1000 )
stocks.columns = ['Bangladesh', 'Pakistan', 'Srilanka', 'India']
```

```python
stocks.cumsum().plot()
```

```python
stocks.to_excel('indiansubcontinent.xlsx')
```

```python
stocks.plot.box()
```

```python
stocks.plot.hist(bins=100)
```

```python
stocks.plot.area(stacked=True)
```

```python
stocks.plot.scatter(x="India", y="Pakistan")
```
